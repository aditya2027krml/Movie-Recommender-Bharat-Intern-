```python
import pandas as pd   #importing pandas library
```

```python
data = pd.read_csv(r'netflix_titles.csv')   #reading the input dataset
```

```python
data.head()
```

PRE-PROCESSING DATA

```python
data['tags'] = data['director'] + '' + data['cast']+ ''+ data['listed_in']+ ''+ data['description']
```

```python
data
```

```python
data = data[['show_id', 'type', 'title', 'tags']]
```

```python
data
```

```python
data.dropna(inplace=True) #removed any null value
```

```python
data
```

```python
data.reset_index(inplace=True, drop=True)  #reseting the index after removing the null value
```

```python
data
```

MODEL CREATION

```python
from sklearn.feature_extraction.text import CountVectorizer
```

CountVectorizer - used to convert the text data into numerical data before passing to the cosine_similarity algorithm

```python
cv = CountVectorizer(max_features=5700, stop_words='english')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

Cosine similarity measures the similarity between two vectors by calculating the cosine of the angle between the two vectors.

```python
vector = cv.fit_transform(data['tags'].values.astype('U')).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

PREDICTING VALUES FROM MODEL

```python
data[data['title']=='Zodiac'].index[0] #taking an example
```

```python
similarity[5696]
```

```python
similar_movies=sorted(list(enumerate(similarity[5696])), reverse=True, key =lambda vector:vector[1])
```

```python
for i in similar_movies[1:6]:                                 #we want only 5 recommended movies
    print(data.iloc[i[0]].title)
    print(i[1])  #percentage of similarity
```

```python
def recommend_movie(movie_name):
    # Check if the movie exists in the DataFrame
    if movie_name not in data['title'].values:
        print(f"The movie '{movie_name}' does not exist in the dataset.")
        return
    
    # Get the index of the movie
    movie_index = data[data['title'] == movie_name].index[0]
    
    # Assuming 'similarity' and 'data' are defined elsewhere in your code
    similar_movies = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda vector: vector[1])
    
    # Print recommendations
    print(f"Recommendations for '{movie_name}':")
    for i in similar_movies[1:6]:
        print(data.iloc[i[0]].title + ' with similarity of ' + str(i[1] * 100)[:5])  # Fixed the parenthesis here

        # Assuming 'i[1]' is the similarity score

# Call the function with the movie name
recommend_movie('Ganglands')
```
